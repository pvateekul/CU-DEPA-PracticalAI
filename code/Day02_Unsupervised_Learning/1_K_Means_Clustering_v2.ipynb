## Method Used

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

```python
import numpy as np
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Silhouette Score 
K = range(2, 10)  # Silhouette score requires k >= 2
silhouette_scores = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data[0])
    score = silhouette_score(data[0], kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Selecting k with the Silhouette Method')
plt.show()

best_k = K[np.argmax(silhouette_scores)]
print(f"Best k: {best_k} with silhouette score: {max(silhouette_scores):.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

def plot_silhouette_subplots(data):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()

    for idx, k in enumerate(range(2, 6)):
        ax = axes[idx]

        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(data)

        silhouette_vals = silhouette_samples(data, labels)
        avg_score = silhouette_score(data, labels)

        y_lower = 0
        for i, cluster_id in enumerate(sorted(np.unique(labels), reverse=True)):
            cluster_vals = np.sort(silhouette_vals[labels == cluster_id])
            size = len(cluster_vals)
            y_upper = y_lower + size

            ax.fill_betweenx(
                np.arange(y_lower, y_upper),
                0,
                cluster_vals,
            )
            ax.text(-0.08, y_lower + size / 2, str(cluster_id),
                    ha='center', va='center', fontsize=10)
            y_lower = y_upper

        ax.axvline(x=avg_score, color='red', linestyle='--', linewidth=1.5)
        ax.set_title(f'k = {k}  |  avg score: {avg_score:.3f}', fontsize=12)
        ax.set_xlabel('Silhouette coefficient')
        ax.set_ylabel('Cluster label')
        ax.set_xlim([-0.1, 1.0])
        ax.set_yticks([])

    plt.suptitle('Silhouette Plots for k = 2 to 5', fontsize=14, fontweight='bold', y=1.01)
    plt.tight_layout()
    plt.show()

plot_silhouette_subplots(data[0])
```
